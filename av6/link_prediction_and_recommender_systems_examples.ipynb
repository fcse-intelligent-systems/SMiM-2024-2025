```python
from torch.optim import SGD
from torch.optim import Adam
from torch.utils.data import DataLoader
from torch_geometric.nn import LightGCN
from torch_geometric.datasets import IMDB
from torch_geometric.transforms import RandomLinkSplit
from lightgcn_recommendation import train
from gnn_link_prediction import Model, train_link_prediction
```

```python
data = IMDB('IMDB')
dataset = data[0]
dataset
```

# Link Prediction with GNN

```python
train_val_test_split = RandomLinkSplit(num_val=0.1,
                                       num_test=0.1,
                                       add_negative_train_samples=True,
                                       edge_types=('movie', 'to', 'actor'),
                                       rev_edge_types=('actor', 'to', 'movie'))
```

```python
train_data, val_data, test_data = train_val_test_split(dataset)
```

```python
train_data
```

```python
val_data
```

```python
test_data
```

```python
model_lp = Model(hidden_channels=128, data=dataset)
```

```python
model_lp
```

```python
optimizer_sgd = SGD(model_lp.parameters(), lr=0.0001)
```

```python
train_link_prediction(model=model_lp,
                      train_data=train_data,
                      val_data=val_data,
                      optimizer=optimizer_sgd,
                      epochs=5)
```

# Recommendation with LightGCN

```python
num_nodes = dataset.num_nodes
num_nodes
```

```python
dataset2 = dataset.to_homogeneous()
```

```python
dataset2
```

```python
dataset2.num_nodes
```

```python
model_lightgcn = LightGCN(num_nodes=dataset2.num_nodes,
                          embedding_dim=128,
                          num_layers=1)
```

```python
data_loader = DataLoader(range(dataset2.edge_index.size(1)),
                         shuffle=True,
                         batch_size=16)
```

```python
optimizer_sgd2 = SGD(model_lightgcn.parameters(), lr=0.01)
```

```python
train(dataset=dataset2,
      train_loader=data_loader,
      model=model_lightgcn,
      optimizer=optimizer_sgd2,
      num_actors=dataset['actor'].num_nodes,
      num_movies=dataset['movie'].num_nodes,
      epochs=5)
```
