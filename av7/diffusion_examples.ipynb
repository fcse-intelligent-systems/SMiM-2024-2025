```python
import networkx as nx
import ndlib.models.epidemics as ep
import ndlib.models.opinions as op
import ndlib.models.ModelConfig as mc
from ndlib.viz.mpl.DiffusionTrend import DiffusionTrend
```

```python
g = nx.erdos_renyi_graph(1000, 0.1)
```

```python
g.nodes
```

```python
g.edges
```

```python
model = ep.SIRModel(g)
```

```python
model2 = ep.SIModel(g)
```

```python
config = mc.Configuration()
```

```python
config2 = mc.Configuration()
```

```python
config.add_model_parameter('beta', 0.001)
config.add_model_parameter('gamma', 0.01)
config.add_model_parameter('fraction_infected', 0.05)
```

```python
config2.add_model_parameter('beta', 0.001)
config2.add_model_parameter('fraction_infected', 0.05)
```

```python
model.set_initial_status(config)
```

```python
model2.set_initial_status(config2)
```

```python
model.iteration()
```

```python
iterations = model.iteration_bunch(200)
```

```python
iterations2 = model2.iteration_bunch(200)
```

```python
trends = model.build_trends(iterations)
```

```python
trends2 = model2.build_trends(iterations2)
```

```python
viz = DiffusionTrend(model, trends)
```

```python
viz2 = DiffusionTrend(model2, trends2)
```

```python
%matplotlib inline
viz.plot()
```

```python
%matplotlib inline
viz2.plot()
```
